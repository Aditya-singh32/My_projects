## Let's train a model to see what we are dealing with:

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_log_error, r2_score
```

```python
X = df_tmp.drop("Visitors",axis=1)
y = df_tmp["Visitors"]

X_tmp_train, X_tmp_test, y_tmp_train, y_tmp_test = train_test_split(X,y,test_size=.2,random_state=42)

model = RandomForestRegressor(n_jobs=-1,random_state=42)
model.fit(X_tmp_train,y_tmp_train)
```

### Let's evaluate the model

```python
model.score(X_tmp_test,y_tmp_test)
```

```python
y_tmp_preds = model.predict(X_tmp_test)
r2_score(y_tmp_test,y_tmp_preds)
```

```python
mean_absolute_error(y_tmp_test,y_tmp_preds)
```

The mean_absolute_error is not that bad.

```python
mean_squared_log_error(y_tmp_test,y_tmp_preds)
```

```python
importance = model.feature_importances_
```

```python
importance
```

```python
plt.figure(figsize=(5,5))
plt.xticks(rotation=90)
plt.bar(x=X_tmp_train.columns,height=importance);
```

### Observations:

---
* As predicted Total and Misc. are the most important features.
* Other than that the rest don't contribute that much.

```python
cross_val_score(model,X_tmp_train,y_tmp_train,cv=5)
```

```python
np.mean(cross_val_score(model,X_tmp_train,y_tmp_train,cv=5))
```

```python
np.mean(cross_val_score(model,X_tmp_train,y_tmp_train,cv=5,scoring="neg_mean_absolute_error"))
```

Symbology Used
---

I see some letters and colors used in some of the tables, what do they mean?
Here is a list of special symbology used in the data;

> M - Means that the data is missing.  This can happen for a variety of reasons including the data did not make a quality check, there was an equipment outage, or even the observer was not available at a manual station.

> T - This means that only a trace of precipitation has fallen.  Trace is defined as less than the smallest measurable amount.  That threshold is below for the different precipitation measurements;

Liquid precipitation (rain, showers) - Less than 0.005"
Snowfall - Less than 0.05"
Snow depth on the ground - Less than 0.5"

Red highlighted temperature - The highest temperature value for that time period
Blue highlighted temperature - The lowest temperature value for that time period
Green highlighted precipitation - The greatest amount of precipitation for that time period
