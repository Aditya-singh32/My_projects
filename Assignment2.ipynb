```python
# Standard Library Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
MaxLittle_Parkinsons_Dataset = pd.read_csv('/content/drive/MyDrive/MaxLittleParkinsons.csv')

MaxLittle_Parkinsons_Dataset.head()
```

```python
#Remove the two aforementioned columns to obtain the predictor columns
feature_names = MaxLittle_Parkinsons_Dataset.columns.difference(['status', 'name'])

X = MaxLittle_Parkinsons_Dataset[feature_names]
y = MaxLittle_Parkinsons_Dataset['status'] # it is a series and if we put another [] then it will become a dataframe.
groups = MaxLittle_Parkinsons_Dataset['name']
```

```python
groups
```

```python
# Assuming 'groups' is a Pandas Series containing strings like 'phon_R01_S01_1'

# Extract the subject number using regex and grouping (with parentheses around the digits)
# The exact below one is regular expression.
extracted_groups = groups.str.extract(r'S(\d+)_')

# Convert to NumPy array
# We can keep it as a Pandas Series, but I wanted to convert to NumPy
groups_numpy = extracted_groups.to_numpy()

# Convert to integer
# We can keep it as strings, but I wanted to convert to int
groups_int = groups_numpy.astype(int)

# Flatten the array
# We can keep it as a 2D array, but I wanted to flatten it to a 1D array.
groups_flattened = groups_int.flatten()

print(groups_flattened)
```

## LeaveOneOut

```python
import numpy as np
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score

feature_names = MaxLittle_Parkinsons_Dataset.columns.difference(['status', 'name'])

X = MaxLittle_Parkinsons_Dataset[feature_names]
y = MaxLittle_Parkinsons_Dataset['status']

loo = LeaveOneOut()

# To store the accuracy
accuracy_scores_list = []

model = KNeighborsClassifier(n_neighbors=1)

for train_index, test_index in loo.split(X):
    # Split the data into training and test sets for the current fold
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Fit the model
    #model = KNeighborsClassifier(n_neighbors=1)
    model.fit(X_train, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test)

    # Calculate and store the mean squared error
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores_list.append(accuracy)

# Calculate the average mean squared error across all iterations
average_accuracy = np.mean(accuracy_scores_list)
print(f'Average Accuracy: {average_accuracy}')
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(neigh , X = X , y = y , cv = loo)
```

```python
np.mean(scores)
```

## LeaveOneGroupOut

```python
from sklearn.model_selection import LeaveOneGroupOut

feature_names = MaxLittle_Parkinsons_Dataset.columns.difference(['status', 'name'])

X = MaxLittle_Parkinsons_Dataset[feature_names]
y = MaxLittle_Parkinsons_Dataset['status']

logo = LeaveOneGroupOut()
logo.get_n_splits(X, y, groups_flattened)

model = KNeighborsClassifier(n_neighbors=1)

accuracy_scores_list = []

for train_index, test_index in logo.split(X,y,groups=groups_flattened):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model.fit(X_train, y_train)
    predictions = model.predict(X_test)

    accuracy = accuracy_score(y_test, predictions)
    accuracy_scores_list.append(accuracy)

np.mean(accuracy_scores_list)
```
